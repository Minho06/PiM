```python
import numpy as np
import matplotlib.pyplot as plt
import math

def plot_performance(variable, output, title, labelY):
    plt.figure()
    plt.plot(variable, output, marker='o')
    plt.title(title)
    plt.xlabel('variable (n)')
    plt.ylabel(labelY)
    plt.grid(True)
    plt.show()



def total_time(size, processor_count, compute_time, size_buffer, beta, alpha):

    time = (size**3 * (compute_time / processor_count) +
            (size**3 / size_buffer) * (2 * beta * (1 / processor_count + 1 / math.sqrt(processor_count))) +
            ((size / size_buffer)**3) * (2 * alpha * (1 + math.sqrt(processor_count))))

    return time

def total_energy(size, processor_count, size_buffer, energyMac, energySend):
    energy = (size**3 * (energyMac / processor_count) + 
             (size**3 / size_buffer) * 3 * energySend)
    
    return energy


#define parameters for both time and energy
processor_dim = 16
N = [2,4,6,8,16,32,64,128,256,512]
processor_count = 10

#define parameters for time
alpha = 10
beta = 5
compute_time = 5
size_buffer = 4
times = []

#define parameters for energy
energySend = 5
energyMac = 5
energy = []
```

```python
times = []
energy = []
for size in N:
    times.append(total_time(size, processor_count, compute_time, size_buffer, beta, alpha))
    energy.append(total_energy(size, processor_count, size_buffer, energyMac, energySend))
plot_performance(N, times, 'sizeOfMatrix', 'Times (s)')
plot_performance(N, energy, 'sizeOfMatrix', 'Energy (J)')
```
